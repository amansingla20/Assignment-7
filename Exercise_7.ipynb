**Read the Colleges dataset and clean all the numeric attributes.
From all marks based attributes, remove % sign from different attributes. In SAT/ACT column, Keep higher value.
From all cost based attributes, Remove $ sign and commas.
Remove any null value in any attribute.**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('colleges.csv')
```

```python
unwanted_character = ['%']

def remove_unwanted_character(data):
  separate_list = list(data.loc['*SAT or ACT'].split('%'))
  another_list = separate_list[0].split('/')
  if len(another_list) == 2:
    if another_list[0] < another_list[1]:
      return another_list[1]
    elif another_list[0] == another_list[1]:
      return another_list[0]
    else:
      return another_list[0]

  elif len(another_list) == 1:
    return another_list[0]  


data['*SAT or ACT'] = data.apply(remove_unwanted_character, axis = 1)
print(data['*SAT or ACT'])
```

```python
def unwanted_dollar_sign(data):
  separate_list = list(data.loc['Total Costs'].split('$'))
  another_list_used = separate_list[1].split(',')
  return '' .join(another_list_used)


data['Total Costs'] = data.apply(unwanted_dollar_sign, axis = 1)
print(data['Total Costs'])
```

```python
data['Non-Need-Based Aid+'] = data['Non-Need-Based Aid+'].replace(np.nan , 0)
data.head()
```
